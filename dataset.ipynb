```python
import pandas as pd
import os
```

```python
data = pd.read_csv('csv\styles.csv', on_bad_lines='skip')
data
```

```python
data.drop(['gender', 'masterCategory', 'articleType', 'baseColour', 
           'season', 'year', 'usage', 'productDisplayName'], axis=1, inplace=True)
for category in data['subCategory']:
  if category not in ['Topwear', 'Bottomwear']:
    data.drop(data[data['subCategory'] == category].index, inplace=True)
data
```

```python
images = pd.read_csv('csv\images.csv', on_bad_lines='skip')
images.head()
for i in range(len(images)):
  if int(images['filename'][i][:-4]) not in data['id'].tolist():
    images.drop(images[images['filename'].index == i].index, inplace=True)
images
```

```python
print(images['filename'].tolist())
print(os.listdir('data\images'))
```

```python
import os
for file in os.listdir('data\images'):
    if file not in images['filename'].tolist():
        path = 'data\images\\'+str(file)
        os.remove(path)
```
